```python
import requests
import bs4
```

```python
# torob
page = requests.get('https://torob.com/price-list/3561/%D8%B4%DB%8C%D8%B1-%D9%84%D8%A8%D9%86%DB%8C-%D9%84%DB%8C%D8%B3%D8%AA-%D9%82%DB%8C%D9%85%D8%AA/?page=1')
```

```python
page
```

```python
html = page.content
```

```python
page_soup = bs4.BeautifulSoup(html,"html.parser")
```

```python
# page_soup
```

```python
tables = list(page_soup.findAll(name='table'))
```

```python
len((tables))
```

```python
target_table = tables[0]
```

```python
tr_tags = list(target_table.find_all('tr'))
```

```python
len(tr_tags)
```

```python
ls = []
for tr in tr_tags:
    td_tags = list(tr.find_all('td'))
    ls.append([i.text for i in td_tags])
#     print('*********************************')
# print(ls)
ls = ls[1:]
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(ls, columns =['ProductName', 'ProductPrice', 'ProductSalers'])
```

```python
df.to_excel(excel_writer='p1.xlsx' ,index=False)
```

```python
import bs4 as bs
import urllib.request
import pandas as pd

source = urllib.request.urlopen('https://torob.com/price-list/3561/%D8%B4%DB%8C%D8%B1-%D9%84%D8%A8%D9%86%DB%8C-%D9%84%DB%8C%D8%B3%D8%AA-%D9%82%DB%8C%D9%85%D8%AA/?page=1').read()
soup = bs.BeautifulSoup(source,'lxml')

table = soup.find('table')
table_rows = table.find_all('tr')

for tr in table_rows:
    td = tr.find_all('td')
    row = [tr.text for tr in td]
    print(row)
```
