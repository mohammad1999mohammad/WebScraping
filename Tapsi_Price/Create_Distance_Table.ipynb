```python
import requests
import json
import numpy as np
import pandas as pd
import time
import datetime
from geopy.distance import great_circle as GRC
```

```python
points = []

for lat in range(35900,35400,-5):
    
    for lon in range(50880,51870,5):
        
        points.append({'latitude': lat/1000 ,'longitude': lon/1000 })
    
df = pd.DataFrame(points)
```

```python
df.to_excel('points.xlsx')
```

```python
all_list = []
for i in range(0,len(input_data),1):
    for item in points:
        my_dict = {}
        my_dict['origin_area'] = input_dictionary['Area_name'][i]
        my_dict['origin_latitude'] = input_dictionary['Source_lat'][i]
        my_dict['origin_longitude'] = input_dictionary['Source_long'][i]
        my_dict['destination_latitude'] = item['latitude']
        my_dict['destination_longitude'] = item['longitude']
        my_dict['distance (km)'] = GRC((my_dict['origin_latitude'],my_dict['origin_longitude']),(my_dict['destination_latitude'],my_dict['destination_longitude'])).km
        
        all_list.append(my_dict)
```

```python
with pd.ExcelWriter("Distance.xlsx".format(datetime.datetime.now().strftime("%X").replace(":", "_" ))) as writer:
    
    df2 = pd.DataFrame(all_list)
    
    df2.to_excel(writer, sheet_name='Distance', index=False)
        
```

```python
df2_less_2 = df2[df2['distance (km)']<=2]
```
